```python
import h5py
import os
import tensorflow as tf
import random
import numpy as np
import seaborn as sns
from tensorflow import keras
from PIL import Image
from matplotlib import pyplot as plt
from matplotlib import image
from sklearn.preprocessing import FunctionTransformer
from sklearn.model_selection import train_test_split
from tensorflow.keras.callbacks import Callback
from tensorflow.keras import backend as K
from math import exp
```

```python
# Load data
h5_file = h5py.File('food_data_224_224_3.h5')

# Grab individual data
images_dset = h5_file.get('images')
categories_dset = h5_file.get('categories')
category_names_dset = h5_file.get('category_names')
num_samples = categories_dset.shape[0]

category_names = np.array(category_names_dset, dtype='str')
```

```python
num_categories = 50
batch_size = 32
indices = np.arange(num_categories * 1000)
categories = np.array(categories_dset)

seed = 42
# Split dataset into training set and lump set of test and validation sets
train_X, lump_X, train_y, lump_y = train_test_split(indices, 
                                                    categories[:num_categories*1000], 
                                                    test_size=0.3, 
                                                    stratify=categories[:num_categories*1000], 
                                                    random_state=seed)

# Split lump set into test set and validation set
test_X, val_X, test_y, val_y = train_test_split(lump_X,
                                                lump_y,
                                                test_size=0.5,
                                                stratify=lump_y,
                                                random_state=seed)

num_train_batches = len(train_X) // batch_size
num_val_batches = len(val_X) // batch_size
num_test_batches = len(test_X) // batch_size
num_warmup_batches = num_train_batches // 5
init_lr = 0.001
```

```python
def food_batch_generator(images_dset, categories_dset, batch_size, num_steps, indices):
    while True:
        num_batches = len(indices) // batch_size
    
        np.random.shuffle(indices)
    
        for batch_num in range(min(num_batches, num_steps)):
            indices_batch = indices[batch_num * batch_size:(batch_num + 1) * batch_size]
            indices_batch.sort()
            images_batch = images_dset[indices_batch]
            categories_batch = categories_dset[indices_batch]
            yield (images_batch, categories_batch)
```

```python
def image_augment(image):
    if random.random() < 0.5:
        image = np.fliplr(image)
    if random.random() < 0.5:
        image = np.flipud(image)
    return image

def food_batch_generator_augment(images_dset, categories_dset, batch_size, num_steps, indices):
    while True:
        num_batches = len(indices) // batch_size
    
        np.random.shuffle(indices)
    
        for batch_num in range(min(num_batches, num_steps)):
            indices_batch = indices[batch_num * batch_size:(batch_num + 1) * batch_size]
            indices_batch.sort()
            images_batch = images_dset[indices_batch]
            categories_batch = categories_dset[indices_batch]
            yield (np.array([image_augment(image) for image in images_batch]), categories_batch)
```

```python
class FoodLearningRateScheduler(Callback):
    def __init__(self, warmup_epochs, total_epochs, init_lr, verbose=0):
        super(FoodLearningRateScheduler, self).__init__()
        self.warmup_epochs = warmup_epochs
        self.total_epochs = total_epochs
        self.init_lr = init_lr
        self.verbose = verbose

    def on_epoch_begin(self, epoch, logs=None):
        if self.params['steps'] <= self.warmup_epochs:
            lr = self.init_lr * (epoch + 1) / self.warmup_epochs
            K.set_value(self.model.optimizer.lr, lr)
            if self.verbose > 0:
                print(f'Batch {epoch + 1}/{self.warmup_epochs}, '
                      f'Learning Rate: {lr:.6f}')
            if self.params['steps'] == self.warmup_epoch:
                self.init_lr = lr
```

```python
# Define the CNN model architecture
model = keras.Sequential([
    keras.layers.Conv2D(400,
                        kernel_size=(8,8),
                        strides=(2,2),
                        activation='relu',
                        input_shape=(128,128,3)),
    keras.layers.MaxPooling2D((7,7),
                               strides=(2,2)),
    keras.layers.Conv2D(100,
                        kernel_size=(4,4),
                        strides=(2,2),
                        activation='relu',
                        input_shape=(28,28,400)),
    keras.layers.MaxPooling2D((6,6),
                              strides=(2,2)),
    keras.layers.Conv2D(50,
                        strides=(1,1),
                        kernel_size=(2,2),
                        activation='relu',
                        input_shape=(4,4,100)),
    keras.layers.Conv2D(50,
                        strides=(1,1),
                        kernel_size=(3,3),
                        activation='relu',
                        input_shape=(3,3,50)),
    keras.layers.Flatten(),
    keras.layers.Dense(128, activation='relu'),
    keras.layers.Dense(128, activation='relu'),
    keras.layers.Dense(num_categories, activation='softmax')
])
```

```python
from keras import applications
model = keras.Sequential([
    keras.layers.InputLayer((224, 224, 3)),
    keras.layers.RandomFlip("horizontal_and_vertical"),
    keras.layers.RandomRotation(0.4),
    keras.layers.RandomCrop()
])

model.add(applications.MobileNetV3Large(input_shape=(224, 224, 3),
                                        include_top=False,
                                        weights='imagenet',
                                        pooling='max'
                                       ))

'''model.add(applications.InceptionV3(input_shape=(224, 224, 3),
                                        include_top=False,
                                        weights='imagenet',
                                        pooling='max'
                                       ))'''

model.layers[-1].trainable = False

model.add(keras.layers.Dense(400, activation='relu'))
model.add(keras.layers.Dense(256, activation='relu'))
model.add(keras.layers.Dense(128, activation='relu'))
model.add(keras.layers.Dense(num_categories, activation='softmax'))
```

```python
model.summary()
```

```python
# Compile the model
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Train the model
history = model.fit(food_batch_generator(images_dset, categories_dset, batch_size, num_train_batches, train_X),
                    steps_per_epoch=num_train_batches,
                    validation_data=food_batch_generator(images_dset, categories_dset, batch_size, num_val_batches, val_X),
                    validation_steps=num_val_batches,
                    epochs=50,
                    callbacks=[
                        FoodLearningRateScheduler(num_warmup_batches, num_train_batches, init_lr)
                    ])
```

```python
# Evaluate the model on the test data
loss, accuracy = model.evaluate(food_batch_generator(images_dset, categories_dset, batch_size, num_test_batches, test_X),
                                steps=num_test_batches)
print('Test accuracy:', accuracy)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
test_X.sort()
test_images = images_dset[test_X]
test_categories = categories_dset[test_X]

predictions = np.argmax(model.predict(test_images), axis=1)

cm = confusion_matrix(test_categories, predictions)
# clr = classification_report(test_categories, predictions, target_names=category_names[0:num_categories], zero_division=0)
```

```python
plt.figure(figsize=(30, 30))
sns.heatmap(cm, annot=True, fmt='g', vmin=0, cmap='Blues', cbar=False)
plt.xticks(ticks=np.arange(num_categories) + 0.5, labels=category_names[0:num_categories], rotation=90)
plt.yticks(ticks=np.arange(num_categories) + 0.5, labels=category_names[0:num_categories], rotation=0)
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()
```

```python
# Use the model to predict food for a new image
new_image = keras.preprocessing.image.load_img('tests/babybackribs1.jpg', target_size=(224, 224))
new_image_array = keras.preprocessing.image.img_to_array(new_image)
new_image_array = np.expand_dims(new_image_array, axis=0)

prediction = model.predict(new_image_array)
plt.imshow(new_image_array[0] / 256)
plt.axis("off")
print(category_names[np.argmax(prediction)])
```
